```python
import pandas as pd
```

```python
df_users1 = pd.read_csv('./address1.csv')
df_users2 = pd.read_csv('./address2.csv')
df_users3 = pd.read_csv('./address3.csv')
df_orders = pd.read_csv('./orders.csv')
```

```python
df_users = pd.concat([df_users1,df_users2,df_users3])
del df_users1, df_users2, df_users3
```

```python
df_users
```

```python
df_users = df_users.drop(columns=['Accounts Address → User ID', 'Accounts Address → Created'])
df_users = df_users.rename(columns={
    'ID': 'UserID',
    'Accounts Address → Detail': 'Address',
    'Regions Area - Area → Name': 'Area',
    'Regions City - City → Name': 'City',
    'Regions Province - Province → Name': 'Province',
    'Accounts Address → Updated': 'Address Updated'
})
df_users
```

```python
df_users = df_users.sort_values(['Address Updated'], ascending=False).groupby('UserID').head(1)
```

```python
df_users = df_users[df_users['Address Updated'].notna()]
```

```python
df_orders = df_orders[
    [
        'Orders Skill - Skill → User ID',
        'Orders Demand - Demand → User ID',
        'Listings Category - Category → Title',
        'Listings Category - Category → Slug'
    ]
]
df_orders = df_orders.rename(columns = 
                            {
                                'Orders Skill - Skill → User ID' :'WorkmanID',
                                'Orders Demand - Demand → User ID':'DemanderID',
                                'Listings Category - Category → Title':'CategoryTitle',
                                'Listings Category - Category → Slug':'CategorySlug'
                            })
df_orders
```

```python
result = pd.merge(df_orders,df_users,how = 'inner',left_on = 'DemanderID',right_on = 'UserID')
result = result.drop(columns = ['UserID'])
result = result.rename(columns = {
    'Address' :'DemanderAddress',
    'Area':'DemanderArea',
    'City':'DemanderCity',
    'Province':'DemanderProvince'
})
result
```

```python
result = pd.merge(result,df_users,how = 'inner',left_on = 'WorkmanID',right_on = 'UserID')
result = result.drop(columns = ['UserID'])
result = result.rename(columns = {
    'Address' :'WorkmanAddress',
    'Area':'WorkmanArea',
    'City':'WorkmanCity',
    'Province':'WorkmanProvince'
})
result
```

```python
result = result.drop(columns=['Address Updated_x', 'Address Updated_y'])
result
```

```python
result_cleaning = result[result['CategorySlug'] == 'cleaning_base_cat']
result_catering = result[result['CategorySlug'] == 'cleaning']
result_moving = result[result['CategorySlug'] == 'moving']
result_cleaning_and_catering = result[(result['CategorySlug'] == 'cleaning') | (result['CategorySlug'] == 'cleaning_base_cat')]
```

```python
result_cleaning.to_csv('./location results/result_cleaning.csv')
result_catering.to_csv('./location results/result_catering.csv')
result_moving.to_csv('./location results/result_moving.csv')
result_cleaning_and_catering.to_csv('./location results/result_cleaning_and_catering.csv')
```
